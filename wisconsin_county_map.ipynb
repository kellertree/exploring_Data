```python
from bs4 import BeautifulSoup
import requests
import folium
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_counties_in_Wisconsin'
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
links = soup.find_all('a')
```

```python
#for link in links:
    #print(link.get('href'))
```

```python
tables = soup.find_all('table')
```

```python
for table in tables:
    # Extrat rows
    rows = table.find_all('tr')
    
    table_data = []
    
    for row in rows:
        # Extract cells
        cells = row.find_all(['td','th'])
        
        row_data = [cell.get_text(strip=True) for cell in cells]
        table_data.append(row_data)
        
# Print the extracted table data
print(table_data)
```

```python
from geopy.geocoders import Nominatim

# Create a map centered around Wisconsin
map_wisconsin = folium.Map(location=[43.7844, -88.7879], zoom_start=7)

# Extract county names from the list
counties_list = ['Adams', 'Ashland', 'Barron', 'Bayfield', 'Brown', 'Buffalo', 'Burnett', 'Calumet', 'Chippewa', 'Clark', 'Columbia', 'Crawford', 'Dane', 'Dodge', 'Door', 'Douglas', 'Dunn', 'Eau Claire', 'Florence', 'Fond du Lac', 'Forest', 'Grant', 'Green', 'Green Lake', 'Iowa', 'Iron', 'Jackson', 'Jefferson', 'Juneau', 'Kenosha', 'Kewaunee', 'La Crosse', 'Lafayette', 'Langlade', 'Lincoln', 'Manitowoc', 'Marathon', 'Marinette', 'Marquette', 'Menominee', 'Milwaukee', 'Monroe', 'Oconto', 'Oneida', 'Outagamie', 'Ozaukee', 'Pepin', 'Pierce', 'Polk', 'Portage', 'Price', 'Racine', 'Richland', 'Rock', 'Rusk', 'Sauk', 'Sawyer', 'Shawano', 'Sheboygan', 'St. Croix', 'Taylor', 'Trempealeau', 'Vernon', 'Vilas', 'Walworth', 'Washburn', 'Washington', 'Waukesha', 'Waupaca', 'Waushara', 'Winnebago', 'Wood']

# Initialize a geocoder
geolocator = Nominatim(user_agent="my_geocoder")

# Iterate over each county and add a marker for it
for county in counties_list:
    # Geocode county location
    location = geolocator.geocode(county + ', Wisconsin')
    if location:
        # Add marker for the county
        folium.Marker([location.latitude, location.longitude], popup=county).add_to(map_wisconsin)

# Display the map
map_wisconsin
```
