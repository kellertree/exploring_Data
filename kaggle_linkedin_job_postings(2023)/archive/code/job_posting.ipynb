The information below explores the most job postings in 2023 at certain specifications. Data was obtained from an open kaggle source at: https://www.kaggle.com/datasets/arshkon/linkedin-job-postings

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import sqlite3 

csv_linkedin_job_postings = r'C:\Users\Vulcan\OneDrive - Madison College\dataSetExplore\kaggle_linkedin_job_posting(2023)\archive\job_postings.csv'
df = pd.read_csv(csv_linkedin_job_postings)
```

```python
# Displays columns in dataset.
print(df.columns)
```

```python
# Creates a clustered bar graph displaying hte top 20 most posted job titles.

top_titles = df['title'].value_counts().nlargest(20)

plt.figure(figsize=(12, 6))
top_titles.plot(kind='bar', color='royalblue')
plt.title('Top 20 Most Posted Job Titles on LinkedIN')
plt.xlabel('Job Title')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Creates a bar graph for top 15 job median salary.

top_titles_salaries = df.groupby('title')['med_salary'].median().nlargest(15)

plt.figure(figsize=(12, 6))
top_titles_salaries.plot(kind='bar', color='royalpurple')
plt.title('Top 15 Most Posted Job Titles with Median Salaries')
plt.xlabel('Job Title')
plt.ylabel('Median Salary')
plt.xticks(rotation=45, ha='right')
plt.show()
```
